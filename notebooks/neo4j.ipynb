```python
import re

def parse_mentions(tweet):
    user_pat = r'@(\w+)'
    mentions = re.findall(user_pat, tweet)
    return mentions

def parse_hashtags(tweet):
    hashtag_pat = r'#(\w+)'
    hashtags = re.findall(hashtag_pat, tweet)
    return hashtags
```

```python
for tweet in result:
    tweet['mentions'] = parse_mentions(tweet['text'])
    tweet['hashtags'] = parse_hashtags(tweet['text'])
```

```python
def create_hashtag_nodes(hashtags):
    query = '''
    UNWIND $hashtags AS tag
    MERGE (h:Hashtag {hashtag: tag})
    SET h.hashtsg = tag
    '''
    
    graph.run(query, hashtags=hashtags)

def connect_tweets_to_hashtags(hashtag_uses):
    query = '''
    UNWIND $hashtags AS ht
    MATCH (h:Hashtag {hashtag: ht.hashtag})
    MATCH (t:Tweet {tweet_id: ht.tweet_id})
    MERGE (t)-[:Has_HT]->(h)
    '''
    graph.run(query, hashtags=hashtag_uses)
```

# Co-occurence

```python
import pandas as pd
from sklearn.preprocessing import MultiLabelBinarizer
from scipy.sparse import csr_matrix
```

```python
tweets = pd.DataFrame(result)
tweets
```

```python
mlb = MultiLabelBinarizer()

hashtags_matrix = mlb.fit_transform(tweets['hashtags'])
```

```python
hashtags_matrix
```

```python
sparse_matrix = csr_matrix(hashtags_matrix)
```

```python
co_occurrence_matrix = sparse_matrix.T.dot(sparse_matrix)
```

```python
co_occurrence_df = pd.DataFrame.sparse.from_spmatrix(co_occurrence_matrix, index=mlb.classes_, columns=mlb.classes_)
```

```python
co_occurrence_df
```

```python
%%time
cypher_query = '''
UNWIND $data AS row
MERGE (h1:Hashtag {name: row.hashtag1})
MERGE (h2:Hashtag {name: row.hashtag2})
MERGE (h1)-[r:COOCCURS_WITH]->(h2)
SET r.weight = row.weight
'''

data = []
for i in range(co_occurrence_df.shape[0]):
    for j in range(i + 1, co_occurrence_df.shape[1]):
        weight = co_occurrence_df.iloc[i, j]
        if weight > 0:
            hashtag1 = co_occurrence_df.index[i]
            hashtag2 = co_occurrence_df.columns[j]
            data.append({'hashtag1':hashtag1, 'hashtag2':hashtag2, 'weight': int(weight)})
```

```python
%%time
graph.run(cypher_query, data=data)
```

# Location sentiment

### Average sentiment of locations

```python
query = '''
MATCH (l:Location)<-[:`Located In`]-(t:Tweet)-[:Has]->(s:Sentiment)
WITH l, 
     CASE s.airline_sentiment WHEN 'positive' THEN 1 WHEN 'neutral' THEN 0 WHEN 'negative' THEN -1 ELSE 0 END AS sentimentScore
RETURN l.name AS location, avg(sentimentScore) AS averageSentiment
ORDER BY averageSentiment
'''

result = graph.run(query)
```

### Average Sentiment of location but filter only locations with at least x amount of tweets connected

```python
query = '''
MATCH (l:Location)<-[:`Located In`]-(t:Tweet)
WITH l, count(t) AS tweetCount
WHERE tweetCount >= 30
MATCH (l)<-[]-(:Tweet)-[:Has]->(s:Sentiment)
WITH l, CASE s.airline_sentiment WHEN 'positive' THEN 1 WHEN 'neutral' THEN 0 WHEN 'negative' THEN -1 ELSE 0 END AS sentimentScore
WITH l, collect(sentimentScore) AS sentimentScores
RETURN l.name AS location, apoc.coll.avg(sentimentScores) AS averageSentiment
ORDER BY averageSentiment
'''

result = graph.run(query)
```

### Sentiment by airline
